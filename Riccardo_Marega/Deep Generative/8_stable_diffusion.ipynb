```python
# # Install necessary libraries
# !pip install transformers
# !pip install diffusers
# !pip install torch
# !pip install accelerate
# !pip install scipy

# Import required libraries
import torch
from diffusers import StableDiffusionPipeline
from PIL import Image
import matplotlib.pyplot as plt
from huggingface_hub import login
login("token")
# Load pre-trained Stable Diffusion model
model_id = "runwayml/stable-diffusion-v1-5"  # You can change the model id if needed.
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)

# Move model to GPU if available for faster inference
pipe = pipe.to("cuda")
```

```python

# Define your prompt for text-to-image generation
prompt = "a sunset day for fishing"
# prompt = "a newborn laying on its bed"

# Generate the image
with torch.no_grad():
    image = pipe(prompt).images[0]

# Display the generated image
plt.imshow(image)
plt.axis("off")  # No axis
plt.show()

# Optionally save the generated image
image.save("generated_image.png")
```
